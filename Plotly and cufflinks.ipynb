```python
df = pd.DataFrame(np.random.randn(100,3), columns=['A', 'B','C'])
df.head()
df['A'] = df['A'].cumsum() + 20
df['B'] = df['B'].cumsum() + 15
df['C'] = df['C'].cumsum() + 15

df.iplot()
```

## Area chart

```python
df.iplot(kind='area')
```

```python
df.iplot(kind='area', fill=True)
```

### Surface chart

```python
df3 = pd.DataFrame({'x':[10, 20, 30, 20, 10], 'y':[10, 20, 30, 20, 10], 'z':[10, 20, 30, 40, 50]})
df3.head()
```

```python
df3.iplot(kind='surface')
```

```python
df3.iplot(kind='surface', color='rdylbu')
```

### Work with cf Datagen

```python
cf.datagen.sinwave(10, 0.25).iplot(kind='surface')
```

### Scatter 3D

```python
cf.datagen.scatter3d(2, 150, mode = 'stocks')
```

```python
cf.datagen.scatter3d(2, 150, mode = 'stocks').iplot(kind='scatter3d', x='x', y='y', z='z')
```

### Spread chart

```python
df[['A','B']].iplot(kind='spread')
```

### Histogram

```python
df.iplot(kind='hist', bins=25, barmode='group', bargap = 0.5)
```

```python
df.iplot(kind='hist', bins=25, barmode='overlay', bargap = 0.5)
```

### Bubble 3D

```python
cf.datagen.bubble3d(5, 4, mode='stocks')
```

```python
cf.datagen.bubble3d(5, 4, mode='stocks').iplot(kind='bubble3d', x='x', y='y',z='z', size='size')
```
